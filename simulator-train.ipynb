```python
### Import file names
center_my_frame_path = df['center_img']
left_my_frame_path = df['left_img']
right_my_frame_path = df['right_img']
steering = df['steering'].astype(np.float32)

# Load image function
import cv2
def load_image_from_path(img_path):
    img_bgr = cv2.imread(img_path)
    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    return img
```

```python
 
# Create generator 
import sklearn    
import pandas as pd
import numpy as np

def generator(df, batch_size=128, infinite_loop=True):
    """
    Inputs: 
    df : dataframe loaded from CSV file containing paths to captured cockpit images and steering angles
        
        # df should be created like this
        df = pd.read_csv('./tropic_data/driving_log.csv', header=None,
                         names=['center_img', 'left_img', 'right_img', 
                                'steering','throttle','brake','speed'])
    
    batch_size : batch size yielded back to call
    infinite_loop : True if we want to call the generator for ever 
    """

    
    num_samples = df['steering'].size
    
    side_steering_bias = np.float32(0.25)
    
    # saturation
    sat_min = np.float32(-4.0)
    sat_max = np.float32(4.0)
    
    while True: # Loop forever so the generator never terminates
        for offset in range(0, num_samples, batch_size):
            batch_samples = df[offset:offset+batch_size]

            images = []
            angles = []
            for index, sample in batch_samples.iterrows():
                # US side of road
                center_image_US = load_image_from_path(sample['center_img'])
                left_image_US = load_image_from_path(sample['left_img'])
                right_image_US = load_image_from_path(sample['right_img'])
                
                center_angle_US = np.float32(sample['steering'])
                left_angle_US = center_angle_US + side_steering_bias
                right_angle_US = center_angle_US - side_steering_bias
            
                right_angle_UK = min(max(left_angle_US - 1.5*side_steering_bias, sat_min), sat_max)
                center_angle_UK = min(max(right_angle_UK - side_steering_bias, sat_min), sat_max)
                left_angle_UK = min(max(center_angle_UK - side_steering_bias, sat_min), sat_max)
                
                images.append(center_image_US)
                angles.append([center_angle_UK, center_angle_US])
                images.append(left_image_US)
                angles.append([left_angle_UK, left_angle_US])
                images.append(right_image_US)
                angles.append([right_angle_UK, right_angle_US])
                
                # UK side of road  - Flipping images and angles
                center_image_UK = cv2.flip(center_image_US,1) 
                left_image_UK = cv2.flip(right_image_US,1)
                right_image_UK = cv2.flip(left_image_US,1)
                
                center_angle_UK = -center_angle_US # Negative steering angle
                left_angle_UK = center_angle_UK + side_steering_bias
                right_angle_UK = center_angle_UK - side_steering_bias
                
                left_angle_US =  min(max(right_angle_UK + 1.5*side_steering_bias, sat_min), sat_max)
                center_angle_US =  min(max(left_angle_US + side_steering_bias, sat_min), sat_max)
                right_angle_US =  min(max(center_angle_US + side_steering_bias, sat_min), sat_max)
                
                images.append(center_image_UK)
                angles.append([center_angle_UK, center_angle_US])
                images.append(left_image_UK)
                angles.append([left_angle_UK, left_angle_US])
                images.append(right_image_UK)
                angles.append([right_angle_UK, right_angle_US])

            # trim image to only see section with road
            X_train = np.array(images)
            y_train = np.array(angles)
            
            ## Use a gain of 1.5 to increase steering agressitivity 
            #y_train = 1.2*np.tanh(1.2*y_train)
            yield sklearn.utils.shuffle(X_train, y_train)
        
        if not infinite_loop:
            break
```

## Robustness improvement over the drive (a novel idea?)
Having observed that my training dataset is quiet low compared to the dataset provided by Udacity, I was interested in finding a way to improve the robustness of the neural network in a minimal way.

A simple idea is to imagine that the road is a potential field on which the car navigates. This concept is well illustrated on this student video : https://www.youtube.com/watch?v=JT8w94n1kWw

In order to maintain the car in the center of the road we need to make sure that the potential field sharply increases when the vehicle is closer to the sides of the road or its orientation is not towards the center of the road.
With our dataset, the easiest way to artificially increase this field gradient is to increase the steering values in the training dataset by muliplying it by a constant factor (or a gain as called in Control Theory).

By appying a gain to the steering dataset the vehicle steers more steering more agressively when approaching dangerously.

# Construct Neural Net Architecture

```python
# Build a model
from keras.models import Sequential, Model
from keras.layers.core import Dense, Activation, Flatten, Dropout, Lambda
from keras.layers import Input
from keras.activations import relu, softmax
from keras.layers.convolutional import Convolution2D, Cropping2D
from keras.layers.pooling import MaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.engine.topology import Merge, merge

model_input = Input(shape=(160, 320, 3))
model_stem = model_input
model_stem = Cropping2D(cropping=((60,20),(0,0)), input_shape=model_stem.get_shape()[1:]) (model_stem)
model_stem = Lambda(lambda x: (x-128.0)/255.0, input_shape=model_stem.get_shape()[1:]) (model_stem)
model_stem = Convolution2D(32, 3, 3, border_mode='valid', subsample=(2, 2))(model_stem)
model_stem = MaxPooling2D(pool_size=(2, 2), strides=None, border_mode='same') (model_stem)
branch_input = Activation('elu')(model_stem)

branch_input = BatchNormalization(epsilon=0.001, mode=0, axis=3, momentum=0.9) (branch_input)

left_branch = Convolution2D(32, 1 , 1,border_mode='same')(branch_input)

right_branch = Convolution2D(32, 7, 1, border_mode='same')  (branch_input)
right_branch = Activation('elu') (right_branch)
right_branch = Convolution2D(32, 1, 7, border_mode='same') (right_branch)
right_branch = BatchNormalization(epsilon=0.001, mode=0, axis=3, momentum=0.9) (right_branch)
right_branch = Activation('elu') (right_branch)

merged = merge([left_branch, right_branch], mode='sum') 

merged = Convolution2D(48, 3, 3, border_mode='valid', activation='elu', subsample=(2, 2))(merged)
merged = Convolution2D(64, 3, 3, border_mode='valid', activation='elu',  subsample=(2, 2))(merged)
merged = BatchNormalization(epsilon=0.001, mode=0, axis=-1, momentum=0.9)(merged)
merged = Flatten()(merged)
merged = Dropout(0.5)(merged)
merged = Dense(128, activation='elu')(merged)
merged = Dense(32)(merged)
merged = Dense(2, activation='tanh')(merged)
merged = Dense(2)(merged)


model =  Model(input=model_input, output=merged) 
```

```python
# Display the model
print(model.summary())

from keras.utils.visualize_util import plot
plot(model, to_file='model.png', show_shapes=True)
```

```python


from sklearn.model_selection import train_test_split

df_no_zero = sklearn.utils.shuffle(df)
#df_no_zero = sklearn.utils.shuffle(df_no_zero)
train_samples, validation_samples = train_test_split(df_no_zero, test_size=0.1)

# compile and train the model using the generator function
train_generator = generator(train_samples, batch_size=128, infinite_loop=True)
validation_generator = generator(validation_samples, batch_size=128, infinite_loop=True)


SAMPLE_EPOCHS = train_samples['steering'].size*3*2 # 3 images per steering angle (left,center,right), 2 sides of road
VAL_SIZE = validation_samples['steering'].size*3*2

print('SAMPLE_EPOCHS:',SAMPLE_EPOCHS)
print('VAL_SIZE:',VAL_SIZE)

# Train the model
# History is a record of training loss and metrics
model.compile(loss='mse', optimizer='adam')
history = model.fit_generator(train_generator, samples_per_epoch=SAMPLE_EPOCHS,
                              validation_data=validation_generator,
                              nb_val_samples=VAL_SIZE, nb_epoch=5)

print(history.epoch)
plt.plot(history.epoch,history.history['loss'],history.epoch,history.history['val_loss'])
plt.legend(('loss','val_loss'))
plt.grid('on')
plt.show()
```
